```python
# Polynomial Regression
# Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Polynomial_Regression/Position_Salaries.csv')
dataset.head(5)
```

```python
# Since the Posion's are already represented by Level variable no need to one hot encode the categorical variable and only taking X as the Independent Variable and y as the Dependent Variable.

X = dataset.iloc[:,1:2]
y = dataset.iloc[:,2]
```

```python
# Not splitting the data into training and test sets as we dont have much data.
```

```python
# Fitting the Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(X,y)
```

```python
# Fitting the Polynomial Regression to the Training set
from sklearn.preprocessing import PolynomialFeatures
```

```python
polyreg = PolynomialFeatures(degree=2)
X_poly = polyreg.fit_transform(X)
X_poly
```

```python
linreg2 = LinearRegression()
linreg2.fit(X_poly, y)
```

```python
# Visualising the Linear Regression Results
plt.scatter(X, y, color="red")
plt.title("Truth or Bluff(Linear Regression)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.plot(X, linreg.predict(X), color="blue")
plt.show()
```

```python
# Visualising the Polynomial Regression Results
plt.scatter(X, y, color="red")
plt.title("Truth or Bluff(polynomial Regression)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.plot(X, linreg2.predict(polyreg.fit_transform(X)), color="blue")
plt.show()
```

```python
# Predicting a new result with Linear Regression
linreg.predict([[6.5]])
```

```python
# Predicting a new result with Polynomial Regression
linreg2.predict(polyreg.fit_transform([[6.5]]))
```
